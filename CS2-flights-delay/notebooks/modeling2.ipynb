```python
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsRegressor
```

```python
input_file_name = '../data/processed/X_arr_train.pickle'
with open(input_file_name, 'rb') as in_file:
    X_train = pickle.load(in_file)

input_file_name = '../data/processed/X_arr_test.pickle'
with open(input_file_name, 'rb') as in_file:
    X_test = pickle.load(in_file)

input_file_name = '../data/processed/y_arr_train.pickle'
with open(input_file_name, 'rb') as in_file:
    y_train = pickle.load(in_file)

input_file_name = '../data/processed/y_arr_test.pickle'
with open(input_file_name, 'rb') as in_file:
    y_test = pickle.load(in_file)
```

```python
ind = X_train.sample(n=10_000).index
filter = X_train.index.isin(ind)
X_train_ = X_train.loc[filter]
y_train_ = y_train[filter.tolist()]
```

```python
np.linspace(0.001, 0.005, 9)
```

```python
from sklearn.model_selection import 

param_distr = {'n_estimators': list(map(int, np.linspace(100, 1000, 10))),
              'learning_rate': np.linspace(0.001, 0.005, 9)}

gbr = GradientBoostingRegressor()
rndSearch_cv = RandomizedSearchCV(estimator = gbr, param_distributions= param_distr, n_iter= 20, n_jobs=-1, random_state=1812, verbose=10)
rndSearch_cv.fit(X_train_, y_train_)
```

```python
rndSearch_cv.best_score_
```

```python
rndSearch_cv.best_params_
```

```python
gbr_best =GradientBoostingRegressor(learning_rate= 0.3, n_estimators=320, random_state=1812, verbose=10)
gbr_best.fit(X_train_, y_train_)
```

```python
gbr_best.train_score_
```

```python
gbr_best.score(X_test, y_test)
```

```python
ind = X_train.sample(n=100_000).index
filter = X_train.index.isin(ind)
X_train_ = X_train.loc[filter]
y_train_ = y_train[filter.tolist()]
print('X train shape:', X_train_.shape)
print('y train shape:', y_train_.shape)
```

```python
ind = X_test.sample(n = 100_000).index
filter_test = X_test.index.isin(ind)
X_test_ = X_test.loc[filter_test]
y_test_ = y_test.loc[filter_test].astype('float64')
print('X test shape:', X_test_.shape)
print('y test shape:', y_test_.shape)
```

```python
knr_model = KNeighborsRegressor(n_neighbors = 400, n_jobs = -1)
knr_model.fit(X_train_, y_train_.astype('float64'))
```

```python
knr_model.score(X_test_, y_test_)
```

```python
param_grid = {'n_neighbors': [260, 280, 300, 320, 340]}
knr_model = KNeighborsRegressor()
grid_CV = GridSearchCV(estimator= knr_model, param_grid = param_grid, n_jobs = -1, verbose = 10)
grid_CV.fit(X_train_, y_train_.astype('float64'))
```

```python
print('Best score:', grid_CV.best_score_)
print('Best parameters', grid_CV.best_params_)
```

```python
grid_CV.score(X_test_, y_test_)
```

```python
knr_model.score(X_test, y_test_)
```
